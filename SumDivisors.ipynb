The function $\sigma(n)$ is constructed by summing the divisors of a given number $n$ - both prime and composite (including 1 and the number $n$ itself). Out of curiousity, I will similar function, $\delta(n)$, summing over prime divisors of $n$. We can parametrize the $\delta$ function over a subset of the natural numbers - the pure multiples of primes - as $(p^m, p \cdot m) \quad \forall p \in \mathbf{P}$. Through basic analysis of this parametrization, we can see that as we vary m, logarithmic streaks arise as subsets of the points of the $\delta$ function. Similarly, as we vary $p$, we get linear streaks within the same subset. Therefore, we can show that the lower bound of the function grows at most logarithmically and the upper bound grows at least linearly, that is to say that $\Omega(\delta(n)) \leq \log(n)$ and $\mathcal{O}(\delta(n)) \geq n$.

```python
import math
import primeModule
primes = primeModule.primes()
import matplotlib.pyplot as plt
```

```python
def delta(n):
    divisors = {}
    primes.primesLt(n)
    if n in primes.primes:
        return n
    for p in primes.primes:
        if p > n:
            break
        multiplicity = 0
        while n % p == 0:
            n /= p
            multiplicity += 1
        if multiplicity:
            divisors[p] = multiplicity
    return sum([p*divisors[p] for p in divisors])
```

```python
x = [n for n in range(2,101)]
y = [delta(n) for n in x]
l = [2*math.log2(n) for n in x]
```

```python
plt.plot(x, y)
plt.plot(x,x)
plt.plot(x,l)
```

```python
d = {}
for i in range(len(x)):
    d[x[i]] = y[i]
```

```python
sublog = [n for n in d if d[n] < 2*math.log2(n)]
print(sublog)
for n in sublog:
    print(f"x: {n}, delta: {d[n]}")
```
